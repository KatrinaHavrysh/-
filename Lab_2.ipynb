```python
import requests
import re
```

**Task 1.**
Write a Python script that, given a Wikipedia URL for a movie,
will extract the following:
* Release date
* Director
* Reception (if available as a percentage rating in ”Reception” section
* Genre
* Bonus points for other data categories

```python
page = requests.get("https://en.wikipedia.org/wiki/Priscilla_(film)")
page.text
```

* Release date


---



```python
re.findall(r"Release date?.*plainlist.*>\n.*(\d{4})", page.text)
```


* Director
---



```python
re.findall(r"Directed by.*?<a.*?>(.*?)</a>", page.text)
```

We don't have Reception

* Genre

---



Так як жанру окремо виділеного немає, тому можна спробувати шукати серед самого тексту(((((

```python
re.findall(r'American (.*?) film', page.text)
```

*  Bonus points for other data categories

---



Running time

```python
re.findall(r'<td class="infobox-data">(\d+ minutes)', page.text)
```
